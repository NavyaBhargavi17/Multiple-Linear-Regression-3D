```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)  # for reproducible results
X1 = np.random.rand(100, 1) * 10  # random numbers for feature 1
X2 = np.random.rand(100, 1) * 10  # random numbers for feature 2
y = 3 * X1 + 2 * X2 + 5 + np.random.randn(100, 1) * 2  # target with some noise
```

```python
X = np.concatenate([X1, X2], axis=1)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
X1_grid, X2_grid = np.meshgrid(np.linspace(0, 10, 50), np.linspace(0, 10, 50))
X_grid = np.c_[X1_grid.ravel(), X2_grid.ravel()]
```

```python
y_pred = model.predict(X_grid)
y_pred_grid = y_pred.reshape(X1_grid.shape)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X1, X2, y, color='blue', label='Actual data')  # scatter of real data
ax.plot_surface(X1_grid, X2_grid, y_pred_grid, color='red', alpha=0.5)  # regression plane

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
ax.set_title('Multiple Linear Regression (3D Plot)')

plt.show()
```
